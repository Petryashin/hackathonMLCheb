```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("./обращения граждан на портале народный контроль.xlsx")
data_two = pd.read_excel("./Просьбы жителей по благоустройству.xlsx")
def sortChuvash(text):
    a = ['ӗ','ӑ','ҫ','Ă','ă','Ĕ','ĕ','Ÿ','ÿ','Ç','ç']
    for i in a:
        if i in str(text):
            return True
    return False
arrIf = np.array([sortChuvash(text) for text in list(data["Текст обращения"])])
print(len(data))
data = data[~arrIf]
# data.index = pd.RangeIndex(1, len(data.index) + 1)
data = data.set_index(np.arange(1, len(data)+1))
data.drop(labels = "id", axis =1, inplace = True)
print(len(data))
data['ids'] = pd.Series([int(i) for i in range(len(data)+1)])
data_two['ids'] = pd.Series([int(i+10000) for i in range(len(data_two)+1)])
```
